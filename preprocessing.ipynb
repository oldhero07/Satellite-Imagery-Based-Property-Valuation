# Preprocessing & EDA

This notebook handles loading data, cleaning it, and preparing it for the multimodal model.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import cv2

# Settings
sns.set_style("whitegrid")
pd.set_option('display.max_columns', None)
```

```python
train_df = pd.read_excel('train.xlsx')
test_df = pd.read_excel('test.xlsx')

print("Train Shape:", train_df.shape)
print("Test Shape:", test_df.shape)
```

```python
# Basic Cleaning
# Check for missing values
print(train_df.isnull().sum())
```

```python
# Check for image availability
IMAGE_DIR = "satellite_images"

def check_image_exists(row_id):
    path = os.path.join(IMAGE_DIR, f"{row_id}.jpg")
    return os.path.exists(path)

train_df['image_exists'] = train_df['id'].apply(check_image_exists)
print(f"Images found for {train_df['image_exists'].sum()} / {len(train_df)} training samples.")
```
